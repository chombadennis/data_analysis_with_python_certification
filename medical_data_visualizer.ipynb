```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### Import data

```python
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
height_sqrd = (df['height']/100) ** 2
bmi = df['weight']/height_sqrd
df['overweight'] = (bmi > 25).astype(int)
       
df.head(10)
```

### Normalizing for df['cholesterol']

```python
df['cholesterol'] = (df['cholesterol'] > 1).astype(int)
df.head(10)
```

### Normalizing for df['gluc']

```python
df['gluc'] = (df['gluc'] > 1).astype(int)
df.head(10)
```

```python
df.set_index('id')
```

### Creating the df_cat Dataframe with selected variables

```python
df_cat = pd.melt(df, id_vars=["cardio", 'age', 'sex', 'height', 'weight', 'ap_hi', 'ap_lo'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

### Renaming the variable column to feature 

```python
df_cat.rename(columns={"variable": "feature"}, inplace=True)
df_cat
```

### Grouping and reformating the data in df_cat to split it by cardio

```python
df_cat_grouped = df_cat.groupby(["cardio", "feature", "value"]).size().reset_index(name="count")
df_cat_grouped
```

```python
sns.catplot(x="feature", y="count", hue="value", kind="bar", data=df_cat_grouped)
plt.title("Categorical Feature Counts by Cardio")
plt.xlabel("Feature")
plt.ylabel("Count")
plt.show()
```

```python
df_heat = df[
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
df_heat
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```
